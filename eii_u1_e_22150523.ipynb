```python
# 1 Establezca una variable independiente y una dependiente
import numpy as np
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/Dess2004/Estadistica/refs/heads/main/mental_health.csv')
df.head()
X = df['Year']
Y = df['Depressive disorders (share of population) - Sex: Both - Age: Age-standardized']



# a) Hipotesis de casualidad: En los aos presentados los desordenes de depresion van creciendo.

# variable independiente: year
# variable dependiente: depressive disorders


# b)  recta de regresin ajustada
import matplotlib.pyplot as plt
plt.scatter(X, Y, color ='purple')
plt.xlabel('examenes cortos')
plt.ylabel('examen final')
plt.title('Grafico de dispersion')
plt.grid()

b1 = SXY / SXX
b0 = np.mean(Y) - b1 * np.mean(X)
print("pendiente: (b1)")
print(b1)
print("intercepto: (b0)")
print(b0)

Yc = b1 * X + b0

plt.plot(X, Yc, "--", color='black')
plt.scatter(X, Y, color='pink')
plt.xlabel('Year')
plt.ylabel('Depressive disorders (share of population) - Sex: Both - Age: Age-standardized')
plt.title('Grafico de dispersion')
plt.grid()
# c) grafica los puntos y la recta de regresin

# d). Calcule el coeficiente de correlacion y determinacion e interprete el resultado.
SXX = np.sum((X - np.mean(X))**2)
SXY = np.sum((X - np.mean(X)) * (Y - np.mean(Y)))
syy = np.sum((Y - np.mean(Y))**2)

r = SXY / np.sqrt(SXX * syy)
print('Coeficiente de correlacion: ', r)

print('Coeficiente de determinacion: ', r**2)

# e)
# el coeficiente de correlacion salio negativo significando que las variables se mueven en direcciones opuestas.
# mientras tanto el coeficiente de determinacion nos indica que el modelo es poco fiable.

# f) Calcula el intervalo de confianza del 95% para _1 y _0 Qu dice el intervalo de confianza a la posibilidad de que _1 sea igual a cero? (
nivel_de_significancia = 0.05
from scipy.stats import t
t = t.ppf(1 - nivel_de_significancia / 2, len(Y) - 2)
se_b1 = np.sqrt(np.sum((Y - Yc)**2) / (len(Y) - 2)) / np.sqrt(SXX)
confianza_b1 = (b1 - t * se_b1, b1 + t * se_b1)
print("Intervalo de confianza para b1: ", confianza_b1)

# Esta muy  cercana a tender a 0.

# g) Realiza el grfico de los residuales: A partir de este grfico, menciona si los datos cumplen con los supuestos para la regresin: linealidad, normalidad en torno a la recta, homoscedasticidad (hay ms, pero con estas nos bastan)
residuales = Y - Yc
plt.figure()
plt.scatter(X, residuales, color='yellow')
plt.xlabel('Year')
plt.ylabel('Depressive disorders (share of population) - Sex: Both - Age: Age-standardized')
plt.title('Grafico de dispersion')
plt.axhline(y = 0, color = "blue", linestyle = "--")

# Linealidad si
# normalidad en torno a la recta varia
# homoscedasticidad si
```
